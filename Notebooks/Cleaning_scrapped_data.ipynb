```python
# Define a function to group body types
def group_body_type(body_type):
    if isinstance(body_type, str):
        body_type = body_type.lower()
        
        if 'suv' in body_type or 'crossover' in body_type or 'electric' in body_type or 'sport utility' in body_type or 'muv' in body_type or 'sportback' in body_type: 
            return 'SUVs and Crossovers'
        elif 'sedan' in body_type or 'coupe' in body_type or 'fastback' in body_type or 'compact car' in body_type or 'compact' in body_type or 'china: ningbo, zhejiang  belarus: barysaw, minsk' in body_type:
            return 'Sedans and Coupes'
        elif 'hatchback' in body_type or 'minivan' in body_type or 'van' in body_type:
            return 'Hatchbacks and Minivans'
        elif 'truck' in body_type or 'pickup' in body_type or 'pick up' in body_type or 'crew cab' in body_type or 'single-cab' in body_type or 'double cab' in body_type or 'quad cab' in body_type:
            return 'Trucks and Pickups'
        elif 'convertible' in body_type or 'sports' in body_type or 'flat body' in body_type or 'sports car' in body_type or 'supercar' in body_type or 'super car' in body_type or 'roadster' in body_type or 'hyper car' in body_type or 'sport car'in body_type:
            return 'Convertibles and Sports Cars'
        elif 'wagon' in body_type or 'wagen' in body_type or 'mpv' in body_type or "compact mpv" in body_type:
            return 'Vans and Wagons'
        elif 'minicar' in body_type or 'microcar' in body_type or 'small' in body_type or 'mini car' in body_type:
            return 'Mini and Micro Cars'
        else:
            return 'Others (Miscellaneous)'
    else:
        return 'Unknown'

# Apply the function to the 'body_type' column
df= df.copy()
df.loc[:, 'body_type_group'] = df['body_type'].apply(group_body_type)
```

```python
# Function to update 'fuel_type_grouped' based on 'engine_type'
def update_fuel_type(row):
    if row['fuel_type_grouped'] == "Other":
        # Ensure 'engine_type' is a string to avoid AttributeError
        engine = str(row['engine_type']).lower()  # Convert to lowercase string
        
        # Use patterns to infer fuel type
        if "phev" in engine:
            return "Plug-in Hybrid"
        elif any(x in engine for x in ["electric", "hydrogen", "motor", "ev", "flux", "fuel cell"]):
            return "Electric"
        elif any(x in engine for x in ["diesel", "mhawk100", "k9k dci"]):
            return "Diesel"
        elif any(x in engine for x in ["gas", "petrol", "tce", "tfsi", "mitsubishi","turbo"]):
            return "Gasoline/Petrol"
        elif "hybrid" in engine:
            return "Hybrid"
        
        
    return row['fuel_type_grouped']  # Keep the original value if it's not "Other"

# Apply the function to update 'fuel_type_grouped'
df['fuel_type_grouped'] = df.apply(update_fuel_type, axis=1)
```

```python
def group_transmission(trans):
    # Convert the value to string if it's a float or NaN
    if isinstance(trans, float) or trans is None:
        return "Unknown"
    
    trans = str(trans).lower()  # Convert to lowercase string

    if "manual" in trans or 'gearbox' in trans:
        return "Manual"
    elif "automatic" in trans and "dual" not in trans and "semi" not in trans or 'utomatic' in trans or 'auto' in trans or 'multi-clutch' in trans:
        return "Automatic"
    elif "cvt" in trans:
        return "CVT"
    elif "dual-clutch" in trans or "dct" in trans or "dual clutch" in trans or 'dsg' in trans or 'isr' in trans:
        return "Dual-Clutch"
    elif ("single-speed" in trans or "electric" in trans or 'direct drive' in trans
            or 'planetary' in trans or 'multi speed' in trans or 'single speed' in trans or "1" in trans ):
        return "Electric"
    else:
        return "Other"

# Apply grouping
df['transmission_group'] = df['transmission_type'].apply(group_transmission)

# Display grouped data
df['transmission_group'].value_counts()
```

```python
def group_drive_type(drive_type):
    if isinstance(drive_type, str):
        drive_type = drive_type.lower()

        # Grouping various drive types into broader categories
        if ('all-wheel' in drive_type or 'awd' in drive_type  or "all wheel drive" in drive_type or 'allwheel drive' in drive_type
            or 'al-wheel drive' in drive_type or "alll-wheel drive" in drive_type):
            return 'All-Wheel Drive (AWD)'
        elif ('4wd' in drive_type or '4x4' in drive_type or 'four wheel drive' in drive_type or 'fourl-wheel drive' in drive_type
              or 'four-wheel drive' in drive_type or 'fron-wheel drive' in drive_type or 'four -wheel drive' in drive_type):
            return 'Four-Wheel Drive (4WD)'
        elif 'front-wheel' in drive_type or 'fwd' in drive_type or 'front wheel drive' in drive_type:
            return 'Front-Wheel Drive (FWD)'
        elif 'rear-wheel' in drive_type or 'rwd' in drive_type or "rear wheel drive" in drive_type or 'real wheel drive' in drive_type or 'rear'in drive_type:
            return 'Rear-Wheel Drive (RWD)'
        else:
            return 'Others (Miscellaneous)'
    else:
        return 'Unknown'

# Apply the function to the 'drive_type' column
# df = df.copy()
df.loc[:, 'drive_type_group'] = df['drive_type'].apply(group_drive_type)

# # Display the grouped drive types
# print(df[['drive_type', 'drive_type_group']])
```

```python
start_time = time.time()
df_missing_year = df[df['year'].isna()]

# Function to scrape the car year from a given URL
def scrape_year_from_url(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url)
        
        # Check if request was successful
        if response.status_code == 200:
            # Parse the page content
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Find the relevant text using the provided CSS selector
            element = soup.select_one('span > strong')

            # If the element is found, extract the text
            if element:
                text = element.get_text()
                
                # Use regex to extract a year starting with '2'
                year_match = re.findall(r'\b2\d{3}\b', text)

                if year_match:
                    return year_match[0]  
                else:
                    return None  
                # print(url, year_match)
            else:
                return None  
        else:
            return None  
    except Exception as e:
        print(f"Error while scraping {url}: {e}")
        return None  

# Iterate through the DataFrame and scrape the missing year for each URL
for index, row in df_missing_year.iterrows():
    url = row['car_link']  
    
    # Scrape the year
    scraped_year = scrape_year_from_url(url)
    
    # Update the DataFrame with the scraped year
    if scraped_year:
        df.loc[index, 'year'] = scraped_year

end_time = time.time()

# Calculate and print the total runtime
runtime = end_time - start_time
print(f"Script completed in {runtime:.2f} seconds.")
```
